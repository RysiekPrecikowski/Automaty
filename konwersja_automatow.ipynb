```python
from FA import FiniteAutomata, TransitionFunction
```

# zadanie 1

Automat akceptuje napisy majÄ…ce na 3 od konca pozycji 1

```python
func1 = TransitionFunction(
    {
        'q1': {'0': {'q1'},
               '1': {'q1', 'q2'}},

        'q2': {'0': {'q3'},
               '1': {'q3'}},

        'q3': {'0': {'q4'},
               '1': {'q4'}},

        'q4': {'0': {},
               '1': {}},
    },
    {'q4'},
    'q1'
)
func_converted1 = func1.convert_to_deterministic()
automata1 = FiniteAutomata(func1)

automata1.run('0110')
automata1.visualizer.show()
```

```python
automata1_converted = FiniteAutomata(func_converted1)

automata1_converted.run('0110')
automata1_converted.visualizer.show()
```
